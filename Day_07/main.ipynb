```python
headers = ['Age','DistanceFromHome','Education','EmployeeCount','EmployeeID','JobLevel','MonthlyIncome','NumCompaniesWorked','PercentSalaryHike','StandardHours','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager']
median = []
mean = []
mode = []
var = []
std = []
kurt = []
skew = []
```

### Median of each column

```python
for i in data_house.median():
    print(i)
    median.append(i)
```

### Mean of each column

```python
for i in data_house.mean():
    print(i)
    mean.append(i)
```

### Mode of each column

```python
data_house[['Age','DistanceFromHome','Education','MonthlyIncome','NumCompaniesWorked', 'PercentSalaryHike','TotalWorkingYears', 'TrainingTimesLastYear','YearsAtCompany','YearsSinceLastPromotion', 'YearsWithCurrManager']].mode()
```

### Variance of each column

```python
for i in data_house.var():
    print(i)
    var.append(i)
```

### Standard Deviation of each column

```python
for i in data_house.std():
    std.append(i)
```

### Kurtosis of each column

```python
for i in data_house.kurt():
    kurt.append(i)
```

### Skewness of each column

```python
data_house.skew()
for i in data_house.skew():
    print(i)
    skew.append(i)
```
